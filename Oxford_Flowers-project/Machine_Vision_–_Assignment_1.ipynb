```python
import os
import json
import shutil
```

**Prepare Dataset into Subfolder**

```python
#Load JSON files containing image-class mapping
json_file_dir1 = '/content/drive/MyDrive/Oxford_Flowers/data/JSON_files/train_set.json'
json_file_dir2 = '/content/drive/MyDrive/Oxford_Flowers/data/JSON_files/val_set.json'

# Path to the train subfolder
train_dataset_dir = '/content/drive/MyDrive/Oxford_Flowers/data/train_subfolder'
val_dataset_dir = '/content/drive/MyDrive/Oxford_Flowers/data/val_subfolder'

#Organize dataset into SubFolder format based on class mapping
image_dir = '/content/drive/MyDrive/Oxford_Flowers/jpg'
```

```python
# Function to check the number of JPEG files mentioned in a JSON file
def count_existing_images(json_file_path, image_dir):
    count = 0
    with open(json_file_path) as f:
        data = json.load(f)
        for images in data.values():
            for image_name in images:
                if os.path.exists(os.path.join(image_dir, image_name)):
                    count += 1
    return count
```

```python
count_existing_images(json_file_dir1, image_dir)
```

```python
# Function to count the number of class names in a JSON file
def count_class_names(json_file_path):
    with open(json_file_path) as f:
        data = json.load(f)
        return len(data)
```

```python
count_class_names(json_file_dir1)
```

```python
# Organize images into subfolders/ TRAIN_SUBFOLDER
with open(json_file_dir1) as f:
    data = json.load(f)

for class_name, images in data.items():
    class_dir = os.path.join(train_dataset_dir, class_name)
    os.makedirs(class_dir, exist_ok=True)
    for image_name in images:
        src = os.path.join(image_dir, image_name)
        if os.path.exists(src):
            dst = os.path.join(class_dir, image_name)
            shutil.move(src, dst)
            print(f"Moved {image_name} to {class_name} folder in flower_subfolder")
        else:
            print(f"Image {image_name} not found in image directory, skipping...")
```

```python
# Organize images into subfolders/ VAL_SUBFOLDER
with open(json_file_dir2) as f:
    data = json.load(f)

for class_name, images in data.items():
    class_dir = os.path.join(val_dataset_dir, class_name)
    os.makedirs(class_dir, exist_ok=True)
    for image_name in images:
        src = os.path.join(image_dir, image_name)
        if os.path.exists(src):
            dst = os.path.join(class_dir, image_name)
            shutil.move(src, dst)
            print(f"Moved {image_name} to {class_name} folder in flower_subfolder")
        else:
            print(f"Image {image_name} not found in image directory, skipping...")
```
